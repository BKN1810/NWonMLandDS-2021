```python
#import packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#read the file
df = pd.read_csv('D:/Dataset/Datamining Dataset/TSLA.csv')
print(df.shape)
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
# setting the index as date
df['Date'] = pd.to_datetime(df.Date,errors='ignore')
df.index = df['Date']
df.head()
```

```python
df.info()
```

```python
#plot
plt.figure(figsize=(16,10))
plt.plot(df['Adj_Close'], label=' Adj Close Price history')
```

```python
# The Simple Moving Average
df['SMA_2'] = df.Adj_Close.rolling(2, min_periods=1).mean()
df['SMA_5'] = df.Adj_Close.rolling(5, min_periods=2).mean()
df['SMA_7'] = df.Adj_Close.rolling(7, min_periods=3).mean()
df['SMA_10'] = df.Adj_Close.rolling(10, min_periods=5).mean()
df.head(7)
```

```python
#plot
plt.figure(figsize=(16,10))
plt.plot(df['Adj_Close'])
plt.plot(df['SMA_2'])
plt.plot(df['SMA_5'])
plt.plot(df['SMA_7'])
plt.plot(df['SMA_10'])
plt.legend(('Adj_Close', 'SMA_2', 'SMA_5','SMA_7','SMA_10'),loc='upper left')
```

```python
plt.figure(figsize=(10,8))
df.iloc[2000:-1].plot(y='SMA_10')
df.iloc[2000:-1].plot(y='SMA_7')
plt.legend(('SMA_7','SMA_10'),loc='upper left')
```

```python
#Commonly used technical indicators include moving averages (SMA: Simple Moving Average , EMA, MACD),
# the Relative Strength Index (RSI), Bollinger Bands (BBANDS), and several others.
# exponential moving average (EMA)
# Add EMA to dataframe by appending 
# Note: pandas_ta integrates seamlessly into our existing dataframe

import pandas_ta
df.ta.ema(close='Adj_Close', length=5, append=False)
df.ta.ema(close='Adj_Close', length=10, append=True)
df.ta.ema(close='Adj_Close', length=20, append=True)
df.head(15)
```

```python
#plot
plt.figure(figsize=(16,10))
plt.plot(df['Adj_Close'])
plt.plot(df['EMA_10'])
plt.plot(df['EMA_20'])
plt.plot(df['EMA_5'])
plt.legend(('Adj_Close', 'EMA_10', 'EMA_20','EMA_5'),loc='upper left')
```

```python
#Count total NaN in each column of Training Data
print(df.isnull().sum())
```

```python
df=df.fillna(df.mean())
# df.fillna(df.mean(), inplace=True)
```

```python
#Count total NaN in each column of Training Data
print(df.isnull().sum())
```

```python
from sklearn.model_selection import train_test_split
# Split data into testing and training sets
X_train, X_test, y_train, y_test = train_test_split(df[['Adj_Close']], df[['EMA_20']], test_size=.2)
```

```python
from sklearn.linear_model import LinearRegression
# Create Regression Model
model = LinearRegression()
# Train the model
model.fit(X_train, y_train)
#predict the value
y_pred=model.predict(X_test)

from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
# Printout relevant metrics
print("Model Coefficients:", model.coef_)
print("Mean Absolute Error:", mean_absolute_error(y_test, y_pred))
print("Coefficient of Determination:", r2_score(y_test, y_pred))
```
